```python
from scipy.stats import t, norm
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Задача 4

```python
mean_array = []
n = 2 #можно задать любое интересующее n для количества случайных величин
rv = t(1)
for k in range(0, 10000):
    x = rv.rvs(size = n)
    mean_array.append(np.mean(x))  

df = pd.DataFrame()
df['mean'] = mean_array
print(df)
sns.histplot(data = df, x = 'mean', stat = 'density')
plt.xlim(-10, 10) #ограничиваем диапазон на оси X, 
                        #чтобы поближе посмотреть график
df2 = pd.DataFrame({'x': np.linspace(-10, 10, 10000)})
df2['PDF'] = rv.pdf(df2['x'])
sns.lineplot(data = df2, x = 'x', y = 'PDF')
```

## Задача 5

```python
rv1 = t(3)
mean_array_1 = []
n = 14 #можно выбрать любое n для количества случайных величин
for k in range(0, 10000):
    x_1 = rv1.rvs(size = n)
    mean_array_1.append(np.mean(x_1))  

df1 = pd.DataFrame()
df1['mean'] = mean_array_1
sns.histplot(data = df1, x = 'mean', stat = 'density')
plt.xlim(-2, 2) #ограничиваем диапазон на оси X, 
                        #чтобы поближе посмотреть график 

df3 = pd.DataFrame({'x': np.linspace(-2, 2, 10000)})
df3['PDF'] = norm.pdf(df3['x'], loc = 0, scale = 9/n)
sns.lineplot(data = df3, x = 'x', y = 'PDF')
```
